```python
ADRIA = open('/home/miquel/ForMiquel/CLEAN/jacobian_list_txt.txt', 'r')

X=np.zeros((4))

for line in ADRIA:
    array = line.split(",")
    array[len(array)-1] = array[len(array)-1].replace("\n","")
    X = np.vstack((X, array))
X=X[1:,:]    
print(X)
```

```python
X2=np.zeros((4))
for s in Subjects:
    rid = s.rid
    for j in s.jacobians:
        step_reference = j.reference.step
        step_target = j.target.step
        label = j.label - 1
        
        if label != -1:
            array = [rid, step_reference, step_target, label]
            X2 = np.vstack((X2, array))
            
X2 = X2[1:,:] 
#X2 = np.sort(X2, axis=0)
print(X2)
        
```
